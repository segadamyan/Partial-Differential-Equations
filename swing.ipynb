# **Partial differential equation (Swing)**
Solution by Sergey Adamyan, using Runge Kutta's method

*O*'' + 0.2*O' + 10*sin*O* = 0

def W = dO/dt

> O(0) = 0.3, W(0) = 0




```python
def runge_kutta(start_time: float, end_time: float, num_of_poinst: int, W_0: int = 0, O_0: int = 0.3):
  f1 = lambda t, O, W: W
  f2 = lambda t, O, W: -0.2*W - 10*np.sin(O)
  V = [(O_0, W_0)]
  step = (end_time - start_time) / num_of_poinst
  T = np.linspace(start_time + step, end_time, num_of_poinst)

  for t in T:
    v_n = V[-1]
    k1 = f1(t, v_n[0], v_n[1]), f2(t, v_n[0], v_n[1])
    k2 = f1(t + step/2, v_n[0] + step/2*k1[0], v_n[1] + step/2*k1[1]), f2(t + step/2, v_n[0] + step/2*k1[0], v_n[1] + step/2*k1[1])
    k3 = f1(t + step/2, v_n[0] + step/2*k2[0], v_n[1] + step/2*k2[1]), f2(t + step/2, v_n[0] + step/2*k2[0], v_n[1] + step/2*k2[1])
    k4 = f1(t + step, v_n[0] + step*k3[0], v_n[1] + step*k3[1]), f2(t + step, v_n[0] + step*k3[0], v_n[1] + step*k3[1])
    v_n_plus_1 = v_n[0] + step/6*(k1[0] + 2*k2[0] + 2*k3[0] + k4[0]), v_n[1] + step/6*(k1[1] + 2*k2[1] + 2*k3[1] + k4[1])
    V.append(v_n_plus_1)
  return V, T
```

```python
V, T = runge_kutta(0, 60, 10000)
```

**Time axis to angle**

```python
plt.plot(T, [v[1] for v in V[:-1]])
```

**Time axis to velocity**

```python
plt.plot(T, [v[0] for v in V[:-1]])
```

**Velocity to angle**

```python
plt.plot([v[0] for v in V[:-1]], [v[1] for v in V[:-1]])
```
